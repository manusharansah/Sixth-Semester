# Generation of Random Number 

```python
# Linear Congruential Generator

m = 16
a = 5
c = 3

seed = 7

x = seed

for i in range(20):
    x = (a * x + c) % m
    r = x / m
    print(f"{i+1:2d}: State={x:2d} Random={r:.4f}")
```

# Chi-Square Test for Random Number Generation


```python
# LCG parameters
m = 16    # modulus
a = 5    # multiplier
c = 3         # increment
x = 7            # seed (X0)

bins = [0] * 10

for _ in range(20):
    x = (a * x + c) % m
    r = x / m
    index = int(r * 10)
    if index == 10:
        index = 9
    bins[index] += 1

print("Observed (O_i):", bins)

# ---- Chi-Square Test ----
n = 20          # total numbers generated
k = len(bins)            # number of bins
expected = n / k  # E_i = 100 for each bin, since uniform distribution assumed

chi_square = 0
for o in bins:
    chi_square += ((o - expected) ** 2) / expected

print("Expected (E_i) per bin:", expected)
print("Chi-square value:", chi_square)
```

## 1. Semantic Network

A **Semantic Network** is a heterogeneous graph that represents knowledge by linking concepts through relationships — similar to how the human brain stores memory.

- **Nodes** → Objects or Concepts (e.g., "Dog", "Animal")
- **Edges** → Relations between them (e.g., "is-a", "has", "lives-in")

**Real-life example:** Google's Knowledge Graph — when you search "Elon Musk", Google shows his company, birthplace, and spouse because they are all connected nodes in a semantic network.

```
[Dog] --is-a--> [Animal]
[Dog] --has--> [Fur]
[Dog] --lives-in--> [House]
```

```python
# Simple semantic network as a dictionary
semantic_net = {
    "Dog": {
        "is-a": "Animal",
        "has": "Fur",
        "lives-in": "House"
    },
    "Animal": {
        "is-a": "LivingBeing"
    }
}

print(semantic_net["Dog"]["is-a"])  # Output: Animal
```

## 2. Part-of-Speech (POS) Tagging

An NLP model **parses each token** and labels it as a grammatical category:
- **Proper Noun** (NNP) — e.g., "Google"
- **Noun** (NN) — e.g., "company"
- **Verb** (VB) — e.g., "builds"
- **Adjective** (JJ) — e.g., "smart"

**Real-life example:** Gmail uses POS tagging to understand "Meeting at 5pm" — it identifies "Meeting" as a noun and "5pm" as a time entity to auto-suggest a calendar event.

```python
import nltk
nltk.download('averaged_perceptron_tagger')

sentence = "Google builds smart AI models"
tokens = nltk.word_tokenize(sentence)
tags = nltk.pos_tag(tokens)

print(tags)
# Output:
# [('Google', 'NNP'), ('builds', 'VBZ'), ('smart', 'JJ'), ('AI', 'NNP'), ('models', 'NNS')]
```

## 3. Named Entity Recognition (NER)

NER identifies and classifies **real-world entities** in text into predefined categories:

| Entity Type  | Example           |
|--------------|-------------------|
| Organization | Google, Tesla     |
| Location     | New York, Nepal   |
| Person       | Elon Musk         |
| Date/Time    | June 2025         |

**Real-life example:** LinkedIn uses NER to extract company names, job titles, and locations from your resume when you upload it — auto-filling your profile fields.

```python
import spacy
nlp = spacy.load("en_core_web_sm")

text = "Elon Musk founded Tesla in California in 2003."
doc = nlp(text)

for ent in doc.ents:
    print(ent.text, "-->", ent.label_)

# Output:
# Elon Musk   --> PERSON
# Tesla       --> ORG
# California  --> GPE
# 2003        --> DATE
```

## 4. Vector Embedding (Word2Vec)

**Vector Embedding** converts words (tokens) into a list of numbers (vectors) so that machines can understand meaning mathematically.

- Similar words have **similar vectors**
- Enables operations like: `King - Man + Woman ≈ Queen`

**Real-life example:** Spotify uses embeddings to find songs similar to your liked tracks — each song is a vector, and nearby vectors mean similar sound/genre.

```python
from gensim.models import Word2Vec

sentences = [
    ["dog", "barks", "loudly"],
    ["cat", "meows", "softly"],
    ["dog", "is", "a", "pet"],
    ["cat", "is", "a", "pet"]
]

model = Word2Vec(sentences, vector_size=5, min_count=1)

# Get vector for "dog"
print(model.wv["dog"])
# Output: [ 0.021  -0.134   0.087   0.045  -0.062 ]

# Find similar words
print(model.wv.most_similar("dog"))
# Output: [('cat', 0.98), ('pet', 0.91), ...]
```

## 5. Knowledge Triples

A **Knowledge Triple** structures information as:

```
Subject --> Predicate --> Object
```

It combines tokens, POS, NER, and embeddings into a machine-readable fact.

**Real-life example:** Wikipedia's structured data (used by Alexa/Siri) stores facts as triples — so when you ask *"Who founded Apple?"*, it looks up `(Apple) --> founded_by --> (Steve Jobs)`.

```python
# Knowledge triple representation
triples = [
    ("Steve Jobs",  "founded",    "Apple"),
    ("Apple",       "located_in", "California"),
    ("iPhone",      "made_by",    "Apple"),
]

# Query: What did Steve Jobs found?
subject_query = "Steve Jobs"
for s, p, o in triples:
    if s == subject_query and p == "founded":
        print(f"{s} founded {o}")
# Output: Steve Jobs founded Apple
```

## 6. Recommendation System (Jaccard Similarity)

Used to measure **how similar two users/items are** based on shared elements.

### Formula

```
Jaccard Similarity = Intersection / Union
```

**Real-life example:** Amazon uses Jaccard similarity to recommend products — if your cart and another user's cart share many items, their other items get recommended to you.

```python
def jaccard_similarity(set_a, set_b):
    intersection = len(set_a & set_b)
    union = len(set_a | set_b)
    return intersection / union

cart_a = {"Banana", "Apple", "Bread"}
cart_b = {"Banana", "Apple", "Orange"}

score = jaccard_similarity(cart_a, cart_b)
print(f"Intersection : {len(cart_a & cart_b)}")   # 2
print(f"Union        : {len(cart_a | cart_b)}")   # 4
print(f"Similarity   : {score:.0%}")              # 50%
```

### Cosine Similarity (bonus)

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# User preference vectors (ratings for 3 products)
user_a = np.array([[1, 1, 0]])   # likes Banana, Apple
user_b = np.array([[1, 1, 1]])   # likes Banana, Apple, Orange

score = cosine_similarity(user_a, user_b)
print(f"Cosine Similarity: {score[0][0]:.2f}")  # 0.82
```
